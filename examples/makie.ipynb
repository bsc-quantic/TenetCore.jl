# Visualization with Makie

TenetCore.jl has support for visualization of Tensor Networks using the GraphMakie.jl library.

```julia
using TenetCore
using GraphMakie
using CairoMakie
using NetworkLayout
using Random
```

```julia
CairoMakie.activate!(type = "svg")
Random.seed!(0)
```

Any type inheriting `AbstractTensorNetwork` automatically implements the `GraphMakie.graphplot` method.

```julia
tn = rand(TensorNetwork, 30, 3)
graphplot(tn; layout=Stress())
```
